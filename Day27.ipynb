# ***Handling Mixed Variable***
Mixed Variable is dataset with:

  * Numerical variables: Age, income, temperature, etc.

  * Categorical variables: Gender, country, occupation, etc.

  * Optionally: Ordinal variables (e.g., education level), binary features, dates, etc.



```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv')
df.head()
```

```python
df['number'].unique()
```

```python
fig=df['number'].value_counts().plot.bar()
fig.set_title('Passanger travelling with')
```

```python
#extracting numerical part
df['number_numerical']=pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
#extracting categorical part
df['number_categorical']=np.where(df['number_numerical'].isnull(),df['number'],np.nan)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['Cabin_num']=df['Cabin'].str.extract('(\d+)') #captures numerical part
df['Cabin_cat']=df['Cabin'].str[0] #captures first letter
df.head()
```

```python
df['Cabin_cat'].value_counts().plot(kind='bar')
```

```python
#extracting the last bit of ticket as number
df['ticket_num']=df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num']=pd.to_numeric(df['ticket_num'],errors='coerce',downcast='integer')

#extracting the first part of ticket as category
df['ticket_cat']=df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat']=np.where(df['ticket_cat'].str.isdigit(),np.nan,df['ticket_cat'])
```
