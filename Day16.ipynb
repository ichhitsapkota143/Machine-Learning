# **Feature Engineering**
image.png \\
Feature Engineering is the process of creating new features (or variables) from raw data to improve the performance of machine learning models. It’s a crucial step in building any predictive model because the quality of features directly influences model accuracy. The goal is to use domain knowledge or data-driven techniques to extract the most relevant information from raw data.

1. Feature Transformation:
To modify existing features to make them more suitable for modeling.

2. Feature Construction:
To create new features by applying transformations or combining existing features to capture more information.

3. Feature Selection: To select the most relevant features while eliminating irrelevant or redundant ones to improve model efficiency.

4. Feature Extraction: To transform raw data into a reduced set of features that still preserve important information.




# **Feature Transformation Methods**
1. Missing Value Imputation
  * Missing values can be filled using statistical measures like mean, median, or mode, or by predicting the missing values using models. This ensures complete datasets for model training.

2. Handling Categorical Values
  * Categorical data is transformed into numerical format using techniques like One-Hot Encoding (binary columns), Label Encoding (integer values), or Target Encoding (mean of target variable by category), making it suitable for machine learning models.

3. Outlier Detection
  * Outliers are identified as values significantly different from the rest of the data, using methods like Z-Score, IQR, or machine learning algorithms (e.g., Isolation Forest), and can be removed or adjusted to improve model performance.

4. Feature Scaling
  * Scaling techniques such as Normalization (Min-Max) or Standardization (Z-Score) adjust the range or distribution of features to ensure they are comparable and improve the performance of models, especially distance-based algorithms.
