# **Column Transformer**
A ColumnTransformer in machine learning is a tool used to apply different preprocessing techniques to different columns of your dataset. It's especially useful when you have a mix of numerical and categorical features that require different preprocessing pipelines (like scaling for numbers and encoding for categories).



```python
import pandas as pd
import numpy as np
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.read_csv('covid_toy.csv')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['has_covid']),df['has_covid'],test_size=0.2)
```

```python
X_train
```

# **Without Using Column Transformer**

```python
#filling(impute) missing/null values of fever class
si=SimpleImputer()
X_train_fever=si.fit_transform(X_train[['fever']])
X_test_fever=si.fit_transform(X_test[['fever']])
X_train_fever
#missing values are replaced by mean value
```

```python
#ordinal encoding of cough column
oe=OrdinalEncoder(categories=[['Mild','Strong']])
#mild is given smaller value, strong is given larger value
X_train_cough=oe.fit_transform(X_train[['cough']])
X_test_cough=oe.fit_transform(X_test[['cough']])
```

```python
#OneHotEncoding for gender and city
ohe=OneHotEncoder(drop='first',sparse_output=False)
X_train_gender_city=ohe.fit_transform(X_train[['gender','city']])
X_test_gender_city=ohe.fit_transform(X_test[['gender','city']])
```

```python
#Extracting age
X_train_age=X_train.drop(columns=['gender','fever','cough','city']).values
X_test_age=X_test.drop(columns=['gender','fever','cough','city']).values
```

```python
X_train_transformed=np.concatenate((X_train_age,X_train_fever,X_train_gender_city,X_train_cough),axis=1)
pd.DataFrame(X_train_transformed)
```

```python
X_test_transform=np.concatenate((X_test_age,X_test_fever,X_test_gender_city,X_test_cough),axis=1)
pd.DataFrame(X_test_transform)
```

# **Same Code using ColumnTransformer**

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer=ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(drop='first',sparse_output=False),['gender','city'])
],remainder='passthrough')
#drop can be used instead of passthrough, which drop the age column
#passthrough ignores the age column
```

```python
transformer.fit_transform(X_train)
pd.DataFrame(transformer.fit_transform(X_train))
```
