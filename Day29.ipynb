# ***Handling Missing Values***

Handling missing data is one of the important tasks that a data scientist needs to be an expert in. When working on real-world projects, it is common to encounter missing data, and dealing with it requires careful planning to avoid bias and ensure accurate analysis and efficient model training.


image.png

# **Complete Case Analysis(CCA)**
Complete Case Analysis, also known as listwise deletion, is a method for handling missing data where only the rows (cases) with no missing values across any of the variables being analyzed are kept. All rows with at least one missing value are removed from the dataset.

# ** Advantages:**
  1. Simplicity: Easy to implement and understand.

  2. No Assumptions: Does not require modeling or imputation of missing values.

  3. Preserves Data Integrity: Only uses observed data, avoiding artificial estimations.

  4. Widely Supported: Most statistical software and methods support CCA by default.

# ** Disadvantages:**
  1. Data Loss: Removes entire rows with any missing value, leading to a smaller dataset.

  2. Reduced Power: Decreases statistical power due to loss of data.

  3. Potential Bias: Can introduce bias if data is not Missing Completely At Random (MCAR).

4. Not Suitable for High Missingness: Ineffective when a large portion of the data is missing.

# ** When to Use CCA:**
  1. Data is Missing Completely at Random (MCAR)
 The likelihood of a data point being missing is unrelated to any observed or unobserved data.

  2. Low Percentage of Missing Data
 Only a small portion of rows have missing values (typically less than 510%).

  3. Simplicity is Preferred
 When a quick analysis is needed and the data loss will not significantly affect the results.

  4. No Strong Dependence on Sample Size
 If the dataset is large enough that removing some rows wont reduce statistical power.
