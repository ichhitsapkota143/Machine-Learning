# **Filtering DataFrame**

```python
ipl.head(2)
```

```python
#find all the final winners
ipl[ipl['MatchNumber']=='Final']['WinningTeam']
```

```python
#step by step
is_final =ipl['MatchNumber']=="Final" #to check weather a match is final or not
final_match=ipl[is_final] #to make a new df of all final matches
final_match[['Season','WinningTeam']]#to get the winning team and season
```

```python
#how many super over finishes have occured
is_superOver=ipl['SuperOver']=='Y'
super_OverMatch=ipl[is_superOver]
count=super_OverMatch['SuperOver'].count()
print(count)
```

```python
ipl[ipl['SuperOver']=='Y'].shape[0]
```

```python
#how many matches has csk won in kolkota
city_check = ipl['City']=='Kolkota'
KolkotaMatches=ipl[city_check]
KolkotaMatches[KolkotaMatches['WinningTeam']=='Chennai Super Kings']['WinningTeam'].count()
```

```python
ipl[(ipl['City']=='Kolkota') & (ipl['WinningTeam']=='Chennai Super Kings')].shape[0]
```

```python
#toss winner is match winner in percentage
wins=ipl[ipl['TossWinner']==ipl['WinningTeam']].shape[0]
total=ipl.shape[0]
percentage_win=wins/total*100
percentage_win
```

```python
#action movies with rating higher than 7.5
movies[(movies['genres']=='Action')&(movies['imdb_rating']>7.5)]
```
