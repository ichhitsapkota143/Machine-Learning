# **Power Transformation**
Power Transformation in machine learning is a technique used to normalize or stabilize the variance of data, especially when features are not normally distributed or have skewed distributions. This transformation makes the data more Gaussian-like, which can improve the performance of many machine learning models.

üîç Why Use Power Transformation?

Many ML algorithms (like linear regression, logistic regression, and LDA) perform better when features are:

  * Normally distributed

  * Homoscedastic (constant variance)

  * Linearly related to the target variable

# **BoxCox Transform**

Only works on strictly positive data (i.e., all feature values > 0).

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import scipy.stats as stats #for plotting QQ plot, to check normal distribution
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score

from sklearn.preprocessing import PowerTransformer
```

```python
df=pd.read_csv('concrete_data.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.describe() #to check negative and zero value
```

```python
x=df.drop(columns=['Strength'])
y=df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# **Applying regression without any transformation**

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
r2_score(y_test,y_pred)
```

```python
#displot without transformation
for col in X_train.columns:
  plt.figure(figsize=(14,4))
  plt.subplot(121)
  sns.histplot(X_train[col],kde=True)
  plt.title(col)

  plt.subplot(122)
  stats.probplot(X_train[col],dist='norm',plot=plt)
  plt.title(col)

plt.show()
```

```python
#no need to scale while using power transform, it automatically applies normalization
```

# **BOX COX Transformation**

```python
pt=PowerTransformer(method='box-cox')
X_train_transformed=pt.fit_transform(X_train+0.000001) #to ignore 0 value
X_test_transformed=pt.fit_transform(X_test+0.000001)

pd.DataFrame({'cols': X_train.columns,'box_cox_lambdas':pt.lambdas_})
```

```python
#applying linear regression on transformed data
lr=LinearRegression()
lr.fit(X_train_transformed,y_train)
y_pred=lr.predict(X_test_transformed)
r2_score(y_test,y_pred)
```

```python
#here we can see that the accuracy has been improved
```

```python
#Using cross val score
pt=PowerTransformer(method='box-cox')
X_transformed=pt.fit_transform(x+0.000001) #to ignore 0 value
lr=LinearRegression()
np.mean(cross_val_score(lr,X_transformed,y,scoring='r2'))
```

```python
#before and after box cox transformation
X_train_transformed_df=pd.DataFrame(X_train_transformed,columns=X_train.columns)
for cols in X_train_transformed_df.columns:
  #before
  plt.figure(figsize=(14,4))
  plt.subplot(121)
  sns.histplot(X_train[cols],kde=True)
  plt.title(cols)

  #after
  plt.subplot(122)
  sns.histplot(X_train_transformed_df[cols],kde=True)
  plt.title(cols)

plt.show()
```

```python
#apply yeo-johnson transform

pt1=PowerTransformer()
X_train_transformed1=pt1.fit_transform(X_train)
X_test_transformed1=pt1.fit_transform(X_test)

lr=LinearRegression()
lr.fit(X_train_transformed1,y_train)
y_pred=lr.predict(X_test_transformed1)
r2_score(y_test,y_pred)
pd.DataFrame({'cols': X_train.columns,'yeo_johnson_lambdas':pt1.lambdas_})
```

```python
#applying cross val score
pt=PowerTransformer()
X_transformed2=pt.fit_transform(x)
lr=LinearRegression()
np.mean(cross_val_score(lr,X_transformed2,y,scoring='r2'))
```

```python
X_train_transformed2_df=pd.DataFrame(X_train_transformed1,columns=X_train.columns)
```

```python
#before and after comparision for Yeo_johnson
for cols in X_train_transformed2_df.columns:
  #before
  plt.figure(figsize=(14,4))
  plt.subplot(121)
  sns.histplot(X_train[cols],kde=True)
  plt.title(cols)

  #after
  plt.subplot(122)
  sns.histplot(X_train_transformed2_df[cols],kde=True)
  plt.title(cols)

plt.show()
```
