# **Encoding Numerical Data into Categorical Data**

# ***Discretization***
Discretization (also called Binning) is the process of converting continuous data or numerical variables into discrete bins or intervals. This is useful in data preprocessing, especially for machine learning models that prefer categorical input or when you want to reduce the effect of minor observation errors.

✅ Why Use Discretization?
  * Reduce noise in data.

  * Handle outliers more effectively.

  * Improve model interpretability.

  * Necessary for certain algorithms (like Naive Bayes).



# **Unsupervised Binning**
This type of binning does not use the target variable. It's purely based on the distribution of the feature (independent variable).


# ***1. Equal Width/ Uniform Binning***
Equal-width binning: Divides the range into bins of equal width.

# 2. Equal Frequency Binning

Equal Frequency Binning, also known as Quantile Binning, divides a continuous variable into bins such that each bin contains (approximately) the same number of observations. It’s an unsupervised method often used to reduce the effect of outliers and to handle skewed distributions.

# ***3. KMeans Binning***

K-Means binning is a form of unsupervised discretization that uses the K-Means clustering algorithm to create bins or intervals for continuous variables. Instead of splitting data by range or frequency, it groups values based on natural clusters in the data.
