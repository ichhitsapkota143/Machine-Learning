# **Relational Plot**


*   to see the statistical relation betweeen 2 or more variables
*   Bivariate analysis \\
Plots under this section:
  1. scatterplot
  2. lineplot


```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
#scatter plot
#it is a axes level function(usually rectangular shape)
sns.scatterplot(tips,x='total_bill',y='tip')
#scatterplot(nameOfData,x=plotOnX-axis,y=plotOnY-axis)
```

```python
#scatterplot in figure level (usually square shaped)
#relplot()
sns.relplot(tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

```python
#it is recommended to use figure level figure than axes level, because we can use figure level figure to obtain different axes plot
```

```python
#hue in scatter
#it is used to plot different category of a variable using different graph
sns.scatterplot(tips,x='total_bill',y='tip',hue='sex')
```

```python
#style is also used to represent different category of a variable using dots, cross
sns.scatterplot(tips,x='total_bill',y='tip',hue='sex',style='time')
```

```python
#size is used to set size of the figure elements, depending upon the numbers
sns.scatterplot(tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
#here 'size' is the entry in the dataset
```

```python
#line plot
#similar to scatter plot, but continuous, usually used for entry against time
sns.lineplot(tips,x='total_bill',y='tip')
```

```python
gap=px.data.gapminder()
temp_df=gap[gap['country']=='Nepal']
temp_df
#importing data from pyplot library
```

```python
#figure level function
sns.relplot(temp_df,x='year',y='lifeExp',kind='line')
```

```python
sns.lineplot(tempor_df,x='year',y='lifeExp',hue='country')
```

```python
#the other difference between figure level and axes level is the ledgend, in figure level it is outside the graph and in axes level it is inside the graph
```

# **Distribution Plot**


*   used for univariate analysis
*   used to find out the distribution
*   Range of the observation
*   Central Tendency



#figure level function
* displot
#axes level function
* histplot
*kdeplot
*rugplot

```python
#plotting univariate histogram
tips
```

```python
#axes level
sns.histplot(tips,x='total_bill')
```

```python
#figure level
sns.displot(tips,x='total_bill',kind='hist')
```

```python
#bins parameter
sns.displot(tips,x='total_bill',kind='hist',bins=20)
#dividing into 20 size
```

```python
sns.displot(tips,x='total_bill',kind='hist',bins=5)
```

```python
#It is also possible to visualize the distribution of categorical variables using the logic of histogram
#Discrete bins are automatically set for categorical variables
sns.displot(tips,x='day',kind='hist')
#count plot
```

```python
#hue parameter
sns.displot(tips,x='tip',kind='hist',hue='sex')
```

```python
#element
sns.displot(tips,x='tip',kind='hist',hue='sex',element='step')
```

```python
titanic=sns.load_dataset('titanic')
titanic
```

```python
sns.displot(titanic,x='age',kind='hist',element='step',hue='sex')
```

```python
#faceting using col and row
sns.displot(tips,x='tip',kind='hist',hue='sex',col='sex')
#this will not work on axes level function
```

```python
#kdeplot
#Kernel Density Estimation (KDE)
#rather than using discrete bins, a kde plot smooths the observation with a Gaussian kernel, producing a continuos density element
sns.kdeplot(tips,x='total_bill')
#gives the density of probability
```

```python
# function level
sns.displot(tips,x='total_bill',kind='kde')
```

```python
#hue parameter
sns.displot(tips,x='total_bill',kind='kde',hue='sex')
```

```python
#fill parameter
sns.displot(tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
#Rugplot
#plots marginal distribution by drawing ticks along the x and y axis
#this function is intended to complement other plots by showing the location of individual obeservation is an unbtrusive way
```

```python
sns.kdeplot(tips,x='total_bill')
sns.rugplot(tips,x='total_bill')
```
