```python
#import the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#2d line plot
#plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
plt.show()
```

```python
#from pandas dataframe
batsman=pd.read_csv('sharma-kohli.csv')
batsman.head()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.show()
```

```python
#label title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit Sharma vs Virat Kolhi')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.show()
```

```python
#colors(hex) and line(width and style) and marker(size)
#colors
plt.plot(batsman['index'],batsman['V Kohli'],color='blue')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red')
plt.title('Rohit Sharma vs Virat Kolhi')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.show()
```

```python
#line style- solid, dashed
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dotted')
#other line styles are dashdot, solid
plt.title('Rohit Sharma vs Virat Kolhi')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.show()
```

```python
#line width
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='dashdot',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',linewidth=2)
plt.title('Rohit Sharma vs Virat Kolhi')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.show()
```

```python
#marker
#line style- solid, dashed
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='dashed',marker='+',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dotted',marker='*')
#other marker are +,-,>,<,o
#markersize is used to specify the size of the marker
plt.title('Rohit Sharma vs Virat Kolhi')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.show()
```

```python
#legend
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='dashed',marker='+',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dotted',marker='*',label='Rohit')
#label is used to give label to legend
plt.title('Rohit Sharma vs Virat Kolhi')
plt.xlabel('Season')
plt.ylabel('Runs')
#plt.show()
plt.legend()#loc='upper right')
#automatically sets the location of legend
```

```python
#limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
plt.show()
```

```python
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
plt.ylim(0,75000) #to trim y axes
plt.xlim(2017,2019) #to trim x axes
plt.show()
#here 45l in 2021 is outlayer value, so to see the appropriate graph
```

```python
#grid
#legend
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='dashed',marker='+',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dotted',marker='*',label='Rohit')
#label is used to give label to legend
plt.title('Rohit Sharma vs Virat Kolhi')
plt.xlabel('Season')
plt.ylabel('Runs')
#plt.show()
plt.legend()#loc='upper right')
plt.grid()
#automatically sets the location of legend
```

# **Scatter Plot**
#used to plot numeric-numeric datas

```python
#plt.scatter simple function
x=np.linspace(-10,10,50)
y = 10*x+3+np.random.randint(0,300,50)
plt.scatter(x,y)
```

```python
#plt.scatter on pandas data
df=pd.read_csv('batter.csv')
df=df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='Red',marker='*')
plt.title('Average vs Strike Rate of Batsman')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.grid()
plt.show()
```

```python
#size
#using seabon to get tips dataset
tips=sns.load_dataset('tips')
tips.head()
```

```python
#s is size used to represent the size of the people
#bigger dots means many people and smaller dots means less people
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*10,color='Red',marker='*')
plt.title('Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.grid()
plt.show()
```

```python
#scatterplot using plt.plot
#size functions cant be used if you plt.plot
#use another parameter 'o'
plt.plot(tips['total_bill'],tips['tip'],'o')
```

# **Bar Chart**

```python
#simple bar chart
children=[10,20,40,10,30]
colors=['red','blue','greeen','yellow','pink']
plt.bar(colors,children,color='Black')
plt.xlabel('Color')
plt.ylabel('Children')
```

```python
#the above bar chart is vertical bar chart
# for horizontal bar char can be optained by barh
children=[10,20,40,10,30]
colors=['red','blue','greeen','yellow','pink']
plt.barh(colors,children,color='Black')
plt.xlabel('Color')
plt.ylabel('Children')
```

```python
#for higher number of categories it is preferred to use horizontala bar graph
```

```python
#Multiple bar chart
df=pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df["2015"])
plt.bar(df['batsman'],df["2016"])
plt.bar(df['batsman'],df["2017"])
#plotting this way is not effective, because one is overlapped by other
```

```python
#multiple bar graph
#xticks is used to give name to each element in the x-axis
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,color='Black')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2016'],width=0.2)
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.title('avg and strike rate of batsman')
plt.xlabel('batsman')
plt.ylabel('avg and strike rate')
plt.show()
```

```python
#a problem
children=[10,20,40,10,45]
colors=['red red red red red red red','blue blue blue blue blue blue blue','green green green green green','yellow yellow yellow yellow yellow yellow','pink pink pink pink pink']
plt.bar(colors,children,color='Black')
plt.xlabel('Color')
#here the x axis label overlaps
```

# **Piechart**

```python
#simple data
data=[23,45,100,20,49]
subjects=['science','maths','stats','english','nepali']
plt.pie(data,labels=subjects)
plt.show()
```

```python
#dataset
df=pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
#colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
#explode and shadow
#remove particular section from the chart
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
