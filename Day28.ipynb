# **Handling Data and Time Variable**
Handling Date and Time variables is crucial when working with datasets that involve time series, event tracking, logs, schedules, or any kind of temporal data.

```python
import pandas as pd
import numpy as np
```

```python
date=pd.read_csv('orders.csv')
time=pd.read_csv('messages.csv')
```

```python
date.head()
```

```python
time.head()
```

```python
date.info()
```

```python
#here date column is an object(string)
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
#now the date column is datetime datatype
```

# **Extracting year**

```python
date['date_year']=date['date'].dt.year
```

```python
date.sample(5)
```

# **Extracting Month**

```python
date['date_month_no']=date['date'].dt.month
```

```python
date['date_month_name']=date['date'].dt.month_name()
```

```python
date.sample(5)
```

# **Extracting Days**

```python
date['date_days']=date['date'].dt.day
```

```python
#day of week
date['date_dow']=date["date"].dt.dayofweek
```

```python
date.sample(5)
```

```python
#dayofweek-name
date['date_dow_name']=date['date'].dt.day_name()
date.drop(columns=['product_id','city_id','orders'],inplace=True)
```

```python
date.sample(5)
```

```python
#to check is weekend?
date['date_is_weekend']=np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0)
```

```python
date.sample(5)
```

```python
#week of year
date['date'].dt.isocalendar()
```

```python
date['date'].dt.isocalendar().week
```

```python
date['date'].dt.quarter
```

```python
#to calculate gaps between two dates(time elapsed between dates)
import datetime
today=datetime.datetime.today()
today
```

```python
today-date['date']
```

```python
(today-date['date']).dt.days
```

```python
#Months passed
np.round((today-date['date'])/np.timedelta64(1,'W'),0)
```
